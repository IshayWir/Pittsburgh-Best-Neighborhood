## The Metric

I used the columns in my dataset titled Greenway, Park Space, and Woodland, and added them up. I then subtracted the Landslide Prone, Undermined, and Flood Plain percentages from that number. Finally, I added the Park Space (Acres per 1000 persons) times .5, so as to give less weight to that number. Doing this allowed me to get a score, the higher, the more greenery, the lower, the greater chance for a land disaster.

## The Best Neighborhood

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('Greenery.csv')

df_sorted = df.sort_values(by='Scores', ascending=True)


plt.figure(figsize=(10, 20))

plt.barh(df_sorted['Neighborhood'], df_sorted['Scores'], 
         color=['#2ecc71' if x > 0 else '#e74c3c' for x in df_sorted['Scores']])

plt.xlabel('Neighborhood Quality Score', fontsize=12)
plt.ylabel('Neighborhood', fontsize=12)
plt.title('Ranking of Neighborhoods: Quality & Safety Score', fontsize=16)
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()

plt.show()
```

What I found to be the best neighborhood was Regent Square. They had the highest score, followed by Highland Park, Allegheny Center, and South Squirrel Hill. The lowest three neighborhoods were Carrick, Knoxville, and Upper Hill.
